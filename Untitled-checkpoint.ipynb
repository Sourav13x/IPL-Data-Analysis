### Ball faced by each batsman

```python
balls=df.groupby(['striker'])['ball'].count().reset_index()
balls
```

### Run Scored by each batsman

```python
runs=df.groupby(['striker'])['runs_off_bat'].sum().reset_index()
runs
```

```python
four=df[df['runs_off_bat']==4]
four
```

```python
runs_4=four.groupby('striker')['runs_off_bat'].count().reset_index()
runs_4.columns=['batsman','4s']
runs_4
```

```python
six=df.groupby('striker')['runs_off_bat'].agg(lambda x: (x==6).sum()).reset_index()
six.columns=['striker','6s']
six
```

```python
player=pd.concat([runs,balls.iloc[:,1],runs_4.iloc[:,1],six.iloc[:,1]],axis=1)
player
```
